```python
# Diviser le texte en phrases en utilisant les points, les points d'exclamation et les points d'interrogation comme délimiteurs
sentences = text.split('.') + text.split('!') + text.split('?')

# Filtrer les éléments vides
sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

# Afficher le nombre total de phrases
print(f"Le document contient {len(sentences)} phrases.")
```
